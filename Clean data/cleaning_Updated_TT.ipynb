```python
import pandas as pd
import numpy as np
uscities = pd.read_csv("cities.csv")
```

```python
uscities_order = ["city", "state_id", "state_name", "population", "density"]
uscities_df = uscities[uscities_order]
uscities_df
```

```python
city_df = uscities_df.groupby(["city"], as_index=False).agg({"state_id": "first", "state_name": "first", "population": "mean", "density": "mean"})
city_df
```

```python
city_new = city_df.rename(columns = {"state_id": "State_ab", "state_name": "State_name"})
city_new
```

```python
city_new.isnull().sum()
```

```python
city_new.shape
```

```python
city_new.info()
```

```python
city_new.to_csv("City_Updated.csv", index=False)
```

```python
income = pd.read_csv("income.csv",engine='python')
income.head()
```

```python
income.shape
```

```python
income_order = ["State_Name", "State_ab", "City", "ALand", "AWater", "Median", "Stdev"]
income = income[income_order]
income.head()
```

```python
# clean Stdev column 
income['Stdev'] = income['Stdev'].map(lambda x: x.lstrip('0+').rstrip('+0i''i'))

# Change Stdev to float 
income['Stdev'] = pd.to_numeric(income['Stdev'])

# Round off the value of Stdev to 2 decimal places 
income.round({"Stdev":2})
```

```python
#income = income.groupby('City').mean().reset_index()
income_df = income.groupby(["City"], as_index=False).agg({"State_Name": "first", "State_ab": "first", "ALand": "mean", "AWater": "mean", "Median": "mean", "Stdev": "mean"})
income_df
```

```python
income_new = income_df.rename(columns = {"City": "city", "State_Name": "State_name"})
income_new
```

```python
income_new.info()
```

```python
income_new.isnull().sum()
```

```python
# Drop the null columns where all values are null
income_new = income_new.dropna(axis='columns', how='all')

# Drop the null rows
income_new = income_new.dropna()
```

```python
income_new.isnull().sum()
```

```python
income_new.info()
```

```python
income_new.to_csv("income_Updated.csv", index=False)
```
